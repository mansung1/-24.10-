```python
# 새로운 데이터셋 로드
tips = sns.load_dataset('tips')
tips.head()
```

```python
fig = plt.figure()
axes = fig.add_subplot(1, 1, 1)
axes.hist(data=tips, x='total_bill', bins = 20) #bin이 데이터의 개수를 요약해주는 파라미터 변수
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_subplot(1, 1, 1)
sns.countplot(data=tips, x='day')  # bin
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_subplot(1, 1, 1)
axes.boxplot(tips['total_bill'])
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_subplot(1, 1, 1)
axes.scatter(tips['total_bill'], tips['tip'])  # 총지불금액과 팁의 관계
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_subplot(1, 1, 1)
axes.boxplot(
    x=[
        tips[tips['sex'] == 'Female']['tip'],
        tips[tips['sex'] == 'Male']['tip']
    ],
    labels=['Female', 'Male']
)
plt.show()
```

```python
# 디변량그래프

colors = {"Female": "#ff1a340", "Male": "#998ec3"}  # 성별에 따른 색상 지정, 여성은 주황색, 남성은 보라색

scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)

axes1.scatter(data=tips,
              x='total_bill',
              y='tip',
              s=tips['size']**2 * 10,  # size를 제곱하여 점의 크기를 키움, s는 점의 크기를 나타내는 매개변수
              c=tips['sex'].map(colors).tolist(),  # map() 결과를 tolist()로 변환하여 전달
              alpha=0.5)  # alpha는 점의 투명도를 나타내는 매개변수

axes1.set_title("Colored by Sex and Sized by Size")
axes1.set_xlabel("Total Bill")
axes1.set_ylabel("Tip")

scatter_plot.suptitle("Total Bill vs Tip")

plt.show()
```

```python
## 일변량 그래프
```

```python
sns.rugplot(data=tips, x='total_bill')
sns.histplot(data=tips, x='total_bill')
sns.kdeplot(data=tips, x='total_bill')
plt.show()
```

```python
# 이변량 그래프

cnt, ax = plt.subplots()
sns.countplot(data=tips, x='day', palette='viridis')
ax.set_title('요일별 건수')
plt.show()
```

```python
cnt, ax = plt.subplots()
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', ax=ax, palette='viridis')
ax.set_title('지불금액과 팁의 관계')
plt.show()
```

```python
cnt, ax = plt.subplots()
sns.regplot(data=tips, x='total_bill', y='tip', ax=ax)  # palette 매개변수 제거, regplot은 하나의 회귀선이 그어짐
ax.set_title('지불금액과 팁의 관계')
plt.show()
```

```python
# joint plot : 한눈에 파악하기 좋은 차트

joint = sns.jointplot(data=tips, x='total_bill', y='tip')
joint.set_axis_labels(xlabel = 'TotalBill', ylabel = 'Tip')
joint.figure.suptitle('TotalBill과 Tips의 조인트차트')
plt.show()
```

```python
kde, ax = plt.subplots()
sns.kdeplot( data=tips, x='total_bill', y = 'tip', fill = True, ax = ax)
ax.set_title("KDE Plot")
plt.show()
```

```python
import numpy as np

# bar-plots
bar, ax = plt.subplots()
sns.barplot(data=tips, x='time', y='total_bill', estimator=np.mean, ax=ax)
ax.set_title("Bar Plot")
plt.show()
```

```python
box, ax = plt.subplots()
sns.boxplot(data=tips, x='time', y='total_bill', ax=ax)
ax.set_title("Box Plot of Total Bill by Time")
plt.show()
```

```python
# violin plot : 이상치로 따로 처리하지 않고 표현해준다는 점에서 차이가 있다. 
bar, ax = plt.subplots()
sns.violinplot(data=tips, x='time', y='total_bill', ax=ax)
ax.set_title("Violin Plot")
plt.show()
```

```python
fig = sns.pairplot(data=tips)
fig.fig.suptitle('Tips 데이터의 관계분석')
plt.show()
```
